```python
import pandas as pd
import numpy as np
```

```python
salary = pd.read_csv("Resources/Salaries.csv")
salary.head(10)
```

```python
salary_year = salary[salary['yearID'] >= 1990]
salary_year.head()
```

```python
salary_year_cleaned = salary_year.drop(columns=['lgID'])
salary_year_cleaned.head()
```

```python
salary_index = salary_year_cleaned.set_index(['playerID', 'yearID'])
salary_index.head()
```

```python
len(salary_index.index)
```

```python
salary_index.dtypes
```

```python
salary_index.dropna()
```

```python
pitching = pd.read_csv("Resources/pitching.csv")
pitching.head()
```

```python
pitching_year = pitching[pitching['yearID'] >= 1990]
pitching_year.head()
```

```python
pitching_index = pitching_year.set_index(['playerID', 'yearID'])
pitching_index.head()
```

```python
len(pitching_index.index)
```

```python
pitching_index.columns
```

```python
pitching_newdf = pitching_index[['teamID', 'G', 'GS', 'H', 'R', 'ER', 'BB', 'SO', 'L','SV', 'W', 'BAOpp', 'ERA']]
pitching_newdf.head()
```

```python
pitching_newdf.dtypes
```

```python
pitching_newdf.dropna()
```

```python
salary_pitching_index = salary_index.reset_index().merge(pitching_newdf, how="right").set_index('playerID')
salary_pitching_index
```
